```python
%matplotlib inline
import shell_model
import astropy.units as u
import shells
import matplotlib.pyplot as plt
import numpy as np
```

## Shell 18

```python
n=17
shell_list = shells.get_shells()
shell = shell_list[n]
model_pars = {
    'dist':414*u.pc, # pc
    'pix_size':7.5*u.arcsec, # arcsec
    'vstep':0.099*u.km/u.s, # km/s
    'acen':shell.ra, # deg
    'dcen':shell.dec, # deg
    'R':0.22*u.pc, # pc
    'dr':0.2*u.pc, # pc
    'vexp':2.5*u.km/u.s, # km/s
    'v0':13.6*u.km/u.s, # km/s
    'samples_per_voxel':27}
```

```python
model_pars['R'] = 0.16*u.pc
model_pars['dr'] = 0.25*u.pc
model_pars['vexp'] = 3*u.km/u.s
model_pars['v0'] = 14*u.km/u.s

fig = shell_model.compare(model_pars=model_pars, normalize=False,
                          contour_levels=30, pv_length_in_radii=5)
```

## Shell 5

```python
n=4
shell = shell_list[n]
#Initial Guesses
model_pars['acen'] = shell.ra.to(u.deg)
model_pars['dcen'] = shell.dec.to(u.deg)
model_pars['R'] = shell.radius.to(u.deg) * model_pars['dist'] * 2*np.pi/(360. * u.deg)
model_pars['vexp'] = (shell.vmax - shell.vmin) / 2.
model_pars['v0'] = (shell.vmax + shell.vmin) / 2.
```

```python
#Parameter experimentation
model_pars['R'] = 0.145 * u.pc
model_pars['dr'] = 0.13 * u.pc
model_pars['vexp'] = 1.25 * u.km / u.s
model_pars['v0'] = 10. * u.km / u.s

fig = shell_model.compare(model_pars=model_pars, normalize=False,
                          contour_levels=20, pv_length_in_radii=5,
                          pv_angle_step=90*u.deg)
```
